```python
#criando -GRÁFICO MOSTRANDO N° DOS CASOS POR ANO

# Agrupando nomes repetidos em um mesmo grupo
grupos = dataset.filtrado.groupby('Country').groups
grupos

# Realizando a soma dos valores associados a cada nome repetido
casos_porPaises= dataset.filtrado.groupby('Country').sum()
casos_porPaises

# Ordenando o DataFrame pelos maiores valores
casos_por_paises_ordenado = casos_porPaises.sort_values(by='Value', ascending=False)
casos_por_paises_ordenado

novo_dataset = pd.merge(casos_por_paises_ordenado , dataset.filtrado[['Country', 'Survey Year']], on='Country', how='left')
novo_dataset
```

```python
import plotly.express as px

# Criando o gráfico de colunas
#grafico_colunas = px.bar(novo_dataset, x='Country', y='Value', title='Casos por País')

# Exibindo o gráfico
#grafico_colunas.show()



# Criando o gráfico de barras com três eixos
grafico_colunas = px.bar(novo_dataset, x='Country', y='Value', color='Survey Year',
                         title='Maior Número de Casos por: Ano e País')

# Ordenando os dados do novo_dataset pela Survey Year
novo_dataset_sorted = novo_dataset.sort_values(by='Survey Year', ascending=False)

# Exibindo o gráfico
grafico_colunas.show()

# Enviando o gráfico para o Chart Studio
pyo.plot(grafico_colunas, filename='grafico_casos_por_pais')
```
